```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("carcrash.csv")
```

```python
dataset.head(20)
```

```python
plt.scatter(dataset['yearacc'],dataset['yearVeh'])
plt.xlabel('Year of Accident')
plt.ylabel('Year of Vehicle')
plt.show()
```

```python
plt.plot(dataset['yearacc'])
plt.plot(dataset['weight'])
plt.xlabel('Year of Accident')
plt.ylabel('weight')
plt.show()
```

```python
sns.scatterplot(x='yearacc',y='yearVeh',data=dataset,)
plt.xlabel('Year of Accident ')
plt.ylabel('Year of Vehicle')
plt.title('Scatter Plot')
plt.show()
```

```python
sns.scatterplot(x='yearacc',y='yearVeh',data=dataset, hue='sex')
plt.title('Scatter Plot')
plt.xlabel('Year of Accident ')
plt.ylabel('Year of Vehicle')
plt.show()
```

```python
sns.lineplot(data=dataset.drop(['yearacc'], axis=1))
plt.title('Lineplot')
plt.show()
```

```python
sns.lineplot(x="yearacc",y="yearVeh",data=dataset,ci=None)
plt.title('Line Plot')
```

```python
sns.distplot(dataset["ageOFocc"])
plt.title('Dist Plot')
```

```python
sns.barplot(data=dataset,x="yearacc",y="ageOFocc",ci=None)
plt.title('Bar Graph')
```

```python
sns.barplot(data=dataset,x="sex",y="weight",hue="frontal")
plt.xlabel('Gender')
plt.title('Bar Graph')
```

```python
sns.countplot(x="sex",data=dataset)
plt.title('Bar Graph')
```

```python
sns.jointplot(x="yearacc",y="yearVeh",data=dataset)
```

```python
sns.boxplot(x="yearacc",y="yearVeh",data=dataset)
plt.title('Box Plot')
```

```python
corr=dataset.corr()
corr 
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
plt.title('Heat Map')
```
